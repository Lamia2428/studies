```python
def plot_corner_points(image, filtered_coords):
    """ plots corners found in image"""

    plt.imshow(image)
    y, x = np.transpose(filtered_coords)
    plt.plot(x, y, 'b.')
    plt.axis('off')
```

```python
def plot_corners(im_lena, im_chess,lena, chess):   
    # display results
    plt.figure(figsize=(8, 3))
    plt.axes([0, 0, 0.3, 0.95])
    plot_corner_points(im_lena, lena)
    plt.axes([0.2, 0, 0.77, 1])
    plot_corner_points(im_chess, chess)

    plt.show()
```
